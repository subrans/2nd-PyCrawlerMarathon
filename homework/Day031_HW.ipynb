## 沒有加 Header 的情況

```python
import requests
r = requests.get('https://www.aicoin.net.cn/')
response = r.text

response
```

## 加上 Header 的情況

```python
import requests
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.aicoin.net.cn/',headers=headers)
response = r.text

response
```
