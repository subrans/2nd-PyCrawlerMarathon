## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
from bs4 import BeautifulSoup
import requests
import random
import pandas as pd
```

```python
proxy_url = "https://free-proxy-list.net/"
proxy_resp = requests.get(proxy_url)
proxy_soup = BeautifulSoup(proxy_resp.text, "html5lib")
```

```python
proxy_soup.find('tbody').find_all('td')
```

```python
proxy_ip = []
proxy_port = []
proxy_https = []
case = 0

for p in proxy_soup.find('tbody').find_all('td'):
    if case == 0:
        proxy_ip.append(p.text)
    elif case == 1:
        proxy_port.append(p.text)
    elif case == 6:
        proxy_https.append(p.text)
    
    case += 1
    if case > 7:
        case = 0
```

```python
print("IP   :", len(proxy_ip), proxy_ip[:5])
print("Port :", len(proxy_port), proxy_port[:5])
print("HTTPS:", len(proxy_https), proxy_https[:5])
```

```python
proxy_ips = []

for i in range(len(proxy_ip)):
    p = "{}:{}".format(proxy_ip[i], proxy_port[i])
    proxy_ips.append(p)
```

```python
print(len(proxy_ips), proxy_ips[:5])
```

```python
for i in range(10):
    ip = random.choice(proxy_ips)
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=10)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip())
    except:
        print('Fail')
```
